```python
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc

path='am_NH'

filen=path+'/test_sh/u.nc'
fu=nc.Dataset(filen,'r')
filen=path+'/test_sh/emc.nc'
femc=nc.Dataset(filen,'r')

u=fu.variables['u'][:,:,0]
lat=fu.variables['lat'][:]
emc=femc.variables['emc'][:,:,0]

filen=path+'/test_sh/u.col.anom.erainterim.nc'
fu=nc.Dataset(filen,'r')
filen=path+'/test_sh/duepy.col.anom.erainterim.nc'
femc=nc.Dataset(filen,'r')

u_lh=fu.variables['u'][:,:,0]
lat_u_lh=fu.variables['lat'][:]
emc_lh=femc.variables['duepy'][:,:,0]
lat_emc_lh=femc.variables['latitude'][:]

print("after reading data")
```

```python
n=1
fig,ax=plt.subplots(figsize=(2,2),dpi=300)
ax.plot(lat,u[n,:],lat_u_lh,u_lh[n,:],lat,lat*0.,'gray')
ax.tick_params(labelsize=5)
plt.legend(["our","LH21"],fontsize=5,loc=9)
ax.set_xlim([-90,0])
ax.set_xticks(np.arange(-90,0.1,15.))
ax.set_ylim([-10,10])
ax.set_yticks(np.arange(-10.,10.1,2.))
ax.grid(lw=0.5,alpha=0.5)
ax.set_xlabel('latitude',fontsize=6)
ax.set_ylabel('',fontsize=6)
ax.set_title('vertical-averaged zonal-mean U anomaly',fontsize=5)

n=366
fig,ax=plt.subplots(figsize=(2,2),dpi=300)
ax.plot(lat,u[n,:],lat_u_lh,u_lh[n,:],lat,lat*0.,'gray')
ax.tick_params(labelsize=5)
plt.legend(["our","LH21"],fontsize=5,loc=9)
ax.set_xlim([-90,0])
ax.set_xticks(np.arange(-90,0.1,15.))
ax.set_ylim([-10,10])
ax.set_yticks(np.arange(-10.,10.1,2.))
ax.grid(lw=0.5,alpha=0.5)
ax.set_xlabel('latitude',fontsize=6)
ax.set_ylabel('',fontsize=6)
ax.set_title('vertical-averaged zonal-mean U anomaly',fontsize=5)

n=731
fig,ax=plt.subplots(figsize=(2,2),dpi=300)
ax.plot(lat,u[n,:],lat_u_lh,u_lh[n,:],lat,lat*0.,'gray')
ax.tick_params(labelsize=5)
plt.legend(["our","LH21"],fontsize=5,loc=9)
ax.set_xlim([-90,0])
ax.set_xticks(np.arange(-90,0.1,15.))
ax.set_ylim([-10,10])
ax.set_yticks(np.arange(-10.,10.1,2.))
ax.grid(lw=0.5,alpha=0.5)
ax.set_xlabel('latitude',fontsize=6)
ax.set_ylabel('',fontsize=6)
ax.set_title('vertical-averaged zonal-mean U anomaly',fontsize=5)

n=1096
fig,ax=plt.subplots(figsize=(2,2),dpi=300)
ax.plot(lat,u[n,:],lat_u_lh,u_lh[n,:],lat,lat*0.,'gray')
ax.tick_params(labelsize=5)
plt.legend(["our","LH21"],fontsize=5,loc=9)
ax.set_xlim([-90,0])
ax.set_xticks(np.arange(-90,0.1,15.))
ax.set_ylim([-10,10])
ax.set_yticks(np.arange(-10.,10.1,2.))
ax.grid(lw=0.5,alpha=0.5)
ax.set_xlabel('latitude',fontsize=6)
ax.set_ylabel('',fontsize=6)
ax.set_title('vertical-averaged zonal-mean U anomaly',fontsize=5)
```
